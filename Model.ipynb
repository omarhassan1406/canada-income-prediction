```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv" ,names = ['years' , 'income'] , skiprows= 1)
df.head()
```

```python
plt.plot(df.years , df.income)
plt.xlabel("Y ears")
plt.ylabel("Income")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['years']] , df.income)
```

```python
reg.predict([[2020]])
```

```python
plt.plot(df.years , df.income)
plt.xlabel("Y ears")
plt.ylabel("Income")
plt.plot(df.years , reg.predict(df[['years']]))
plt.legend(["Real Values", "Predicted Values"])
```

```python
d = pd.read_csv("canada_per_capita_income.csv" , names = ['years' , 'income'] , skiprows= 1)
del d['income']
d.head(3)
```

```python
p = reg.predict(d)
```

```python
d['predicted_income'] = p
```

```python
d.head(3)
```

```python
d.to_csv("predicted income.csv" , index = False)
```
